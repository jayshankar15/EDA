# Handling the missing data #

```python
data['USE'] = data['USE'].fillna(data['USE'].mode)
data['CAR TYPE'] = data['CAR TYPE'].fillna(data['CAR TYPE'].mode)
data['CITY'] = data['CITY'].fillna(data['CITY'].mode)
```

```python
# Checking the outliers for numerical columns before the imputation part

def boxplots(col):
    sns.boxplot(data[col])
    plt.show()

for i in list(data.select_dtypes(exclude=['object']).columns)[0:]:

    boxplots(i)    

```

```python
median1 = data['TRAVEL TIME'].median()
median2 = data['MILES CLOCKED'].median()
mean1 = data['CAR AGE'].mean()
```

```python
data['TRAVEL TIME'] = data['TRAVEL TIME'].replace(np.nan, median1)
data['MILES CLOCKED'] = data['MILES CLOCKED'].replace(np.nan , median2)
data['CAR AGE'] = data['CAR AGE'].replace(np.nan, mean1)
```

# Outliers tretment

```python
def remove_outliers(col):
    sorted(col)
    Q1, Q3 =  col.quantile([0.25, 0.75])
    IQR = Q3 - Q1
    Lower_range = Q1 - 1.5 * IQR
    Upper_range = Q3 + 1.5 * IQR
    return Lower_range, Upper_range
```
